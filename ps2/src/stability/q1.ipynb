```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x0_a, x1_a, y_a, c='b', marker='o')
ax.set_xlabel('x0_a Label')
ax.set_ylabel('x1_a Label')
ax.set_zlabel('y_a Label')
plt.title("dataset A")
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x0_b, x1_b, y_b, c='r', marker='o')
ax.set_xlabel('x0_b Label')
ax.set_ylabel('x1_b Label')
ax.set_zlabel('y_b Label')
plt.title("dataset B")
plt.show()
fig.savefig("sep_b")
```

```python
sep_a = sns.scatterplot(x0_a, x1_a, hue = y_a).get_figure()
```

```python
separate_b = sns.scatterplot(x0_b, x1_b, hue = y_b).get_figure()
```
